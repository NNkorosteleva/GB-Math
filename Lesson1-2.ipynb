Урок 1. Задание 4. Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
x = np.linspace(0,10,121)
y1 = np.cos(x)
y2 = np.cos(3*x)
plt.axis([0, 10, -1, 1])
plt.plot(x, y1, x, y2)
plt.legend(('k=1','k=3'),loc='upper right')
plt.xlabel('x')
plt.ylabel('cos(kx)')
```

Урок 2. Задание 1. Напишите код на Python, реализующий расчет длины вектора, заданного его координатами

```python
def len_x(x1,x2,x3):
    l_x = np.sqrt(x1**2+x2**2+x3**2)
    return l_x
```

Урок 2. Задание 3. Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
x = []
y1 = []
y2 = []
r=5
for i in range(-100,101):
    x_= r*i/100
    x.append(x_)
    y1.append(np.sqrt(r**2-x_**2))
    y2.append(-np.sqrt(r**2-x_**2))   

fig, ax = plt.subplots(1)
ax.plot(x,y1,'-b',x,y2,'-b')
plt.axis([-r, r, -r, r])   
ax.set_aspect(1)
plt.title('Окружность')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = []
y1 = []
y2 = []
a=2
b=8
for i in range(-100,101):
    x_= a*i/100
    x.append(x_)
    y1.append(np.sqrt(1-x_**2/a**2)*b)
    y2.append(-np.sqrt(1-x_**2/a**2)*b)
plt.axis([-a, a, -b, b])    
plt.plot(x,y1,'-b',x,y2,'-b')
plt.title('Эллипс')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = []
y1 = []
y2 = []
p=10
for i in range(1000):
    x_= i/1000
    x.append(x_)
    y1.append(np.sqrt(2*p*x_))
    y2.append(-np.sqrt(2*p*x_))   
plt.plot(x,y1,'-b',x,y2,'-b')
plt.title('Гипербола')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Урок 2. Задание 5. Нарисуйте трехмерный график двух параллельных плоскостей

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-10,10,0.5)
Y = np.arange(-10,10,0.5)
X, Y = np.meshgrid(X,Y)
Z1 = 2*X + 5*Y
Z2 = 2*X + 5*Y + 100
ax.plot_wireframe(X,Y,Z1)
ax.plot_wireframe(X,Y,Z2)

show()
```

Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-10,10,0.5)
Y = np.arange(-10,10,0.5)
X, Y = np.meshgrid(X,Y)
Z1 = np.sqrt(100*X**2 + Y**2)
Z2 = -(X**2 + Y**2)
ax.plot_wireframe(X,Y,Z1)
ax.plot_wireframe(X,Y,Z2)
show()
```
